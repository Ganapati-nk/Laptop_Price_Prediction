```python


# Your DataFrame 'df'
# Assuming 'Memory' column exists in your DataFrame 'df'

# Remove '.0' from the 'Memory' column, if any
df['Memory'] = df['Memory'].astype(str).replace('\.0', '', regex=True)

# Extract numeric part from 'Memory' column
df['Memory_Size'] = df['Memory'].str.extract(r'(\d+)').astype(float)

# Define a function to extract storage type from 'Memory' column
def extract_storage_type(memory_string):
    if 'SSD' in memory_string:
        return 'SSD'
    elif 'HDD' in memory_string:
        return 'HDD'
    elif 'Hybrid' in memory_string:
        return 'Hybrid'
    elif 'Flash Storage' in memory_string:
        return 'Flash Storage'
    else:
        return 'Unknown'

# Apply the function to create a new column for storage type
df['Storage_Type'] = df['Memory'].apply(extract_storage_type)

# Calculate total memory for each storage type
df['HDD'] = df.apply(lambda row: row['Memory_Size'] if row['Storage_Type'] == 'HDD' else 0, axis=1)
df['SSD'] = df.apply(lambda row: row['Memory_Size'] if row['Storage_Type'] == 'SSD' else 0, axis=1)
df['Hybrid'] = df.apply(lambda row: row['Memory_Size'] if row['Storage_Type'] == 'Hybrid' else 0, axis=1)
df['Flash Storage'] = df.apply(lambda row: row['Memory_Size'] if row['Storage_Type'] == 'Flash Storage' else 0, axis=1)

# Drop intermediate columns
df.drop(columns=['Memory_Size', 'Storage_Type'], inplace=True)

df
```

```python
X = df.drop(columns=['Price'])
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.15,random_state=2)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
from sklearn.ensemble import RandomForestRegressor
```

### Random Forest

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,7,10,11])
],remainder='passthrough')

step2 = RandomForestRegressor(n_estimators=100,
                              random_state=3,
                              max_samples=0.5,
                              max_features=0.75,
                              max_depth=15)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
```

### Exporting the Model

```python
import pickle
pickle.dump(pipe,open('pipe.pkl','wb'))
pickle.dump(df,open('df.pkl','wb'))
```
